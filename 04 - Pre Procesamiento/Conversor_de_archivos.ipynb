### Función que segmenta las opiniones consultivas para su preprocesamiento.

```python
import os
import csv

def extract_pdf_text(pdf_reader, start_page, end_page, output_file):
    # Paso 1: Crear una lista para almacenar las páginas
    judgesVotes = []

    # Paso 2: Iterar sobre las páginas del PDF y almacenarlas en la lista
    for page_num in range(start_page, end_page + 1):
        page = pdf_reader.pages[page_num]
        judgesVotes.append(page)

    # Crear la carpeta si no existe
    output_folder = 'ArchivosSegmentados'
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Paso 3: Abre el archivo CSV en modo de escritura y escribe el texto extraído
    output_path = os.path.join(output_folder, output_file)
    with open(output_path, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)

        # Itera sobre las páginas en judgesVotes
        for page in judgesVotes:
            text = page.extract_text()

            # Imprime el texto extraído
            print(text)
            
            # Escribe el texto extraído en el archivo CSV
            writer.writerow([text])
```
